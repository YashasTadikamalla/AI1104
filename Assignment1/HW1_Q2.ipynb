```python
##############################################
# Program to implement K-means algorithm
# HW - Q2
##############################################
# Import all required libraries
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
# Input k and p
k=int(input("ENTER k (the number of clusters) : "))
p=int(input("ENTER p (the dimensions of input data) : "))

# number of points in each cluster
NUM_PTS = 1500

# Selecting means for generating the clusters
means=np.random.randint(0,((3*k)/4)+1,size=(k,p)) # generating a kxp matrix, of random numbers in range [0,(3*k)/4]

# Selecting covariance for generating the clusters
cov=[[0 for i in range(p)] for j in range(p)]
for i in range(p):
  for j in range(p):
    if (i==j):
      cov[i][j]=0.1
    else:
      cov[i][j]=0

# Generating k clusters of p dimensional points, using multinormal distribution
Y = np.random.multivariate_normal(means[0], cov, NUM_PTS).T
for i in range(1,k):
  y = np.random.multivariate_normal(means[i], cov, NUM_PTS).T
  Y = np.concatenate((Y,y), axis = 1)
```

```python
# Implementing K means algorithm

# Setting stopping condition 
epsilon = 0.001

# Initializing error to a large value
error = 10000

# Initializing a list clusters, to hold all the points, categorized into a particular cluster
clusters=[[]for i in range(k)]

# Selecting covariance for generating the centroids
cov=[[0 for i in range(p)] for j in range(p)]
for i in range(p):
  for j in range(p):
    if (i==j):
      cov[i][j]=1
    else:
      cov[i][j]=0

# Initialize centroids, using multinormal distribution
D = np.random.multivariate_normal(np.mean(Y,axis=1), cov, k).T # generated using mean of all the input points
```

```python
# Initializing iteration count to 0
count = 0

while (error > epsilon):
    # Initializing cluster size to 0
    num=[0]*k

    # Initializing centroid update to 0
    cenup=[[0 for i in range(k)] for j in range(p)]

    # Updating clusters based on distance
    for idx in range(k*NUM_PTS):
        cur_y = Y[:,idx]

        # Finding Euclidean distance of current point from each centroid
        distance_vector=[0]*k
        for i in range(k):
          for j in range(p):
            distance_vector[i]+=np.square(cur_y[j]-D[j][i])
          distance_vector[i]=np.sqrt(distance_vector[i])

        # Finding closest centroid 
        min_idx = distance_vector.index(np.min(distance_vector))

        for i in range(p):
          cenup[i][min_idx]+=cur_y[i]
        num[min_idx]+=1

        # Categorizing points, into clusters
        clusters[min_idx].append(cur_y)
    
    # Calculating centroids    
    for i in range(k):
      for j in range(p):
        if(num[i]!=0):
          cenup[j][i]=cenup[j][i]/num[i]

    # Compute distance between centroids
    error=0
    for i in range(k):
      for j in range(p):
        error+=np.square(cenup[j][i]-D[j][i])
    error = np.sqrt(error)
    
    # Updating centroids
    for i in range(k):
      for j in range(p):
        if(num[i]!=0):
          D[j][i]=cenup[j][i]

    # Updating count
    count += 1
    
    # Printing error
    print ('Iteration :', count, 'Error :', error)

# Print centroids
print('\nPredicted Cluster Centroids : \n', D.T)
print('\nActual Cluster Centroids : \n',means)

# for printing points in each cluster, use the following code
# for i in range(k):
#   print('\ncluster ',i+1,' : ',clusters[i])
```
