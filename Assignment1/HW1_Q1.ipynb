```python
##############################################
# Program to implement linear regression
# HW1 - Q1
##############################################
# Import all required libraries
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Generating training data

# Number of training samples
N = 10
# Generate equispaced floats in the interval [0, 2π]
x = np.linspace(0, 2*np.pi, N)
# Generate noise
mean = 0
std = 0.05
# Generate some numbers from the sine function
y = np.sin(x)
# Add noise
y += np.random.normal(mean, std, N)

# reshaping the arrays explicitly for easier computations
x=x.reshape(N,1)
y=y.reshape(N,1)

# plotting the training data
plt.plot(x,y,'x')
plt.axis('equal')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('training data (N = 10)')
plt.show()
```

```python
# Generating testing data

# Number of testing samples
T = int(N/2) # assuming (number of training data) : (number of testing data) = 66.67 : 33.33

# Generate equispaced floats in the interval [0, 2π]
xx = np.linspace(0, 2*np.pi, T)
# Generate noise
mean = 0
std = 0.05
# Generate some numbers from the sine function
yy = np.sin(xx)
# Add noise
yy += np.random.normal(mean, std, T)

# reshaping the arrays explicitly for easier computations
xx=xx.reshape(T,1)
yy=yy.reshape(T,1)

# plotting the testing data
plt.plot(xx,yy,'gx')
plt.axis('equal')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('testing data (T = 5)')
plt.show()
```

```python
# part (a)
# y = b0 + b1x model

# preparing the training input data, by preparing a Nx2 matrix
temp=np.array([1]*N)
temp=temp.reshape(N,1)
X = np.concatenate((temp,x),axis=1)

# preparing the training output data, by preparing a Nx1 matrix
Y = y

# Y = X.beta is the linear regression model.
# calculating optimal beta matrix using closed fom solution
beta=np.dot((np.dot((np.linalg.inv(np.dot(np.transpose(X),X))),np.transpose(X))),Y)

# preparing the testing input data, by preparing a Tx2 matrix
temp=np.array([1]*T)
temp=temp.reshape(T,1)
XX = np.concatenate((temp,xx),axis=1)

# calculating estimated output, using the above calculated beta matrix and XX
YY = np.dot(XX,beta)

print("part(a)\n")
print("Qualitative Performance Report")

# plotting the predicted outputs
plt.plot(xx,YY,'r+')
plt.axis('equal')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('predicted output values (T = 5)')
plt.show()

# calculating MSE: Mean Squared Error
MSE = (np.sum(np.power((YY-yy),2)))/T
print("Quantitative Performance Report")
print("MSE = ",MSE)
```

```python
print("This model is \"underfitting\" for the data, as both training and testing errors are high")
plt.plot(xx,YY,'r+',label='predicted label')
plt.plot(xx,yy,'gx',label='ground truth label')
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('ground truth and predicted labels (T = 5)')
plt.show()
```

```python
# part (b)
# y = b0 + b1x + b2x^2 + .. + bmx^m model

# m = 2
m = 2

# preparing the training input data, by preparing a Nx2 matrix
temp=np.array([1]*N)
temp=temp.reshape(N,1)
X = np.concatenate((temp,x),axis=1)
for i in range(2,m+1):
  X = np.concatenate((X,np.power(x,i)),axis=1)

# preparing the training output data, by preparing a Nx1 matrix
Y = y

# Y = X.beta is the linear regression model.
# calculating optimal beta matrix using closed fom solution
beta=np.dot((np.dot((np.linalg.inv(np.dot(np.transpose(X),X))),np.transpose(X))),Y)

# preparing the testing input data, by preparing a Tx2 matrix
temp=np.array([1]*T)
temp=temp.reshape(T,1)
XX = np.concatenate((temp,xx),axis=1)
for i in range(2,m+1):
  XX = np.concatenate((XX,np.power(xx,i)),axis=1)

# calculating estimated output, using the above calculated beta matrix and XX
YY = np.dot(XX,beta)

print("part(b) & m = 2\n")
print("Qualitative Performance Report")

# plotting the predicted outputs
plt.plot(xx,YY,'r+')
plt.axis('equal')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('predicted output values (T = 5)')
plt.show()

# calculating MSE: Mean Squared Error
MSE = (np.sum(np.power((YY-yy),2)))/T
print("Quantitative Performance Report")
print("MSE = ",MSE)
```
